# Diatom and Flagellate Dynamics Under Storm Forcing in the SoG (2011 vs. 2018)

## Chapter 1: Introduction



### 1.1 Scientific Background and Motivation 

Recent studies have demonstrated that large-scale climate indices, such as the North Pacific Gyre Oscillation (NPGO), significantly influence the physiology and phenology of marine plankton. While it is established that positive NPGO phases (cold years) generally favor diatom production through enhanced nutrient availability, the specific mechanistic links transferring these basin-scale climate signals to local ecosystem dynamics remain underdefined. This study addresses this gap by shifting focus from interannual statistical correlations to the synoptic-scale physical processes that directly control water column stability and nutrient injection.



### 1.2 Study Area: The Strait of Georgia 

The Central Strait of Georgia (SoG) represents a complex estuarine system defined by the interplay between strong tidal mixing, wind forcing, and substantial freshwater discharge from the Fraser River. These physical drivers create a highly dynamic environment where water column stratification competes with vertical mixing, determining the availability of light and nutrients for primary producers. Understanding the spatial and temporal variability of these forces is essential for predicting biological responses in this semi-enclosed sea.




### 1.3 Research Objectives 

The primary objective of this report is to elucidate how along-strait wind events modulate seawater motion and subsequently drive the biological response of diatoms and flagellates. Specifically, this study aims to decompose the dominant modes of sea surface salinity variability using Principal Component Analysis (PCA), establish the correlation between these physical modes and nutrient stoichiometry (Nitrate vs. Silicate), and quantify the biological response to storm events during the contrasting summer conditions of 2011 and 2018.

## Chapter 2: Methodology


### 2.1 Model Description 
We utilized the SalishSeaCast model, a high-resolution, three-dimensional coupled biophysical model based on the NEMO physical engine and the SMELT biological module. This model resolves the complex topography of the Salish Sea and accurately simulates the coupling between physical circulation—including tides, river plumes, and wind-driven currents—and the lower trophic levels, making it an ideal tool for investigating process-oriented mechanisms in the SoG.

### 2.2 Decomposition of Physical Drivers 

To quantify the atmospheric forcing, the wind field was decomposed into along-strait (approximately 305°) and cross-strait components. We applied Principal Component Analysis (PCA) to the modeled Sea Surface Salinity (SSS) field to extract the dominant spatial patterns (EOFs) and temporal evolution (PCs) of salinity variability. These principal components were then statistically correlated with Fraser River discharge rates and wind stress data to attribute specific physical drivers—such as river plumes, wind sloshing, and coastal upwelling—to the observed salinity modes.



### 2.3 Biological Analysis and Case Study Selection 

Biological response was evaluated by analyzing changes in the biomass ratio of diatoms to flagellates and the limiting nutrient concentrations before, during, and after storm events. Two distinct storm cases were selected from July 2011 and July 2018 to represent different background climate states. The analysis focuses on the short-term evolution of the water column vertical structure and the resulting phytoplankton succession during these high-energy synoptic events.

## Chapter 3: Physical Control of Salinity and Nutrients



### 3.1 Characteristics of Wind and River Forcing 

Statistical analysis of the forcing fields reveals that wind energy in the Central SoG is predominantly aligned with the along-strait direction, significantly exceeding the magnitude of cross-strait winds. This directional dominance suggests that wind-driven circulation and sloshing are the primary mechanisms for mechanical energy input, while the Fraser River discharge acts as the primary source of buoyancy, creating a fundamental physical antagonism between wind-induced mixing and river-induced stratification.

### 3.2 Dominant Modes of Salinity Variability 

The PCA results successfully separated the variability of Sea Surface Salinity into three distinct modes. The first mode, explaining the majority of the variance, is strongly correlated with the Fraser River discharge, representing the freshwater plume extension. The second and third modes show significant correlation with along-strait wind events, capturing the signatures of wind-driven sloshing and coastal upwelling. These modes provide a quantitative framework for distinguishing between buoyancy-driven and wind-driven changes in the surface layer.



### 3.3 Coupling Physics to Nutrient Stoichiometry 

A critical finding of this section is the distinct biogeochemical signature associated with each physical mode. The river-dominated mode is characterized by low salinity, strong stratification, and high silicate concentrations but relatively low nitrate levels, creating conditions that favor flagellates. In contrast, the wind-driven modes are associated with increased surface salinity and the entrainment of deep, nitrate-rich water, thereby resetting the surface nutrient ratio to favor rapid diatom growth.

## Chapter 4: Case Studies of Storm Impacts



### 4.1 Physical Response to the Storm Events

 Analysis of the 2011 2018 storm reveals a classic sequence where strong along-strait winds caused asymmetrical physical responses in the two banks and two ends of the strait through upwelling/downwelling and sloshing. 



### 4.2 Biological Response to the Storm Events 

This physical disturbance resulted in a significant injection of nitrate into the euphotic zone, triggering a distinct post-storm diatom bloom. The model shows that the ecosystem was significantly impacted. Upwelling regions shifted from a flagellate-dominated state to a diatom-dominated state in the days following the wind event, confirming the role of wind mixing as a "reset" mechanism for phytoplankton succession. Comparing the 2011 and 2018 events illustrates that while the mechanism of wind-driven supply is consistent, the magnitude of the biological response is sensitive to the pre-existing stability of the environment.



### 4.3 Spatial Heterogeneity of Response 

The biological response to wind events was not uniform across the Strait. Areas directly under the influence of the Fraser River plume exhibited a damped response due to the intense halocline, which resisted wind mixing. Conversely, regions prone to upwelling or located at the periphery of the plume showed the strongest correlation between wind stress and diatom biomass increase, highlighting the spatial modulation of climate signals by local hydrography.

## Chapter 5: Discussion and Conclusion



### 5.1 Mechanisms Linking Climate to Plankton 

This study synthesizes the results to propose a conceptual model where the seasonal ecology of the SoG is defined by the tension between river-induced stratification (silicate-rich, nitrate-poor) and wind-induced mixing (nitrate-rich). We argue that the "cold-phase" advantage described in large-scale climate literature is mechanistically realized through frequent or intense synoptic wind events that disrupt the river plume, alleviate nitrate limitation, and maintain the ecosystem in a diatom-favorable state.



### 5.2 Conclusion 

In conclusion, the SalishSeaCast model confirms that along-strait winds are the primary driver of vertical nutrient replenishment in the Central Strait of Georgia during summer. By isolating the physical modes of salinity, we demonstrated that wind-driven upwelling and sloshing are critical for counteracting riverine stratification. These physical interruptions are essential for sustaining diatom populations, suggesting that changes in the frequency of summer storms are the direct pathway through which climate variability alters the marine food web.
