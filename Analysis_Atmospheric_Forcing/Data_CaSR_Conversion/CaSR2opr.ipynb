| CaSR variable description | CaSR units | HRDPS variable name | HRDPS long name | HRDPS units | Notes |

| Forecast: Corrected U-component of the wind along West-East direction at 10 m | kts | u_wind | U-Component of Wind | m/s  | Same wind component, but units differ: CaSR uses knots, HRDPS uses m/s. |
| Forecast: Specific humidity at 1.5 m | kg kg**-1 | qair | Specific Humidity | kg/kg  | Units are effectively equivalent. |
| Forecast: Air temperature at 1.5 m | deg_C | tair | Temperature | K  | Same temperature variable, but units differ: CaSR uses Celsius, HRDPS uses Kelvin. |
| Forecast: Sea level pressure at surface | mb | atmpres | Pressure Reduced to MSL | Pa  | Same sea-level pressure concept, but units differ: CaSR uses millibars, HRDPS uses pascals. |
| Forecast: Surface incoming infrared flux at surface | W m**-2 | therm_rad | Downward Long-Wave Rad. Flux | W/m^2  | Incoming infrared flux corresponds to downward long-wave radiation. |
| Forecast: Quantity of precipitation model at surface | m | precip | Total Precipitation | kg/m^2  | Same precipitation-related variable, but units differ. For water, 1 kg/m² is approximately 1 mm water equivalent. |
| Forecast: Downward solar flux at surface | W m**-2 | solar | Downward Short-Wave Radiation Flux | W/m^2  | Downward solar flux corresponds to downward short-wave radiation. |
