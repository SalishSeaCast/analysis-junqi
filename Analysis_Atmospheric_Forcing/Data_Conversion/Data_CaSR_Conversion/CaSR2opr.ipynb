|---|---:|---|---|---:|---|
| Forecast: Corrected V-component of the wind along South-North direction at 10 m | kts | v_wind | V-Component of Wind | m/s  | Same wind component, but units differ: CaSR uses knots, HRDPS uses m/s. |
| Forecast: Corrected U-component of the wind along West-East direction at 10 m | kts | u_wind | U-Component of Wind | m/s  | Same wind component, but units differ: CaSR uses knots, HRDPS uses m/s. |
| Forecast: Specific humidity at 1.5 m | kg kg**-1 | qair | Specific Humidity | kg/kg  | Units are effectively equivalent. |
| Forecast: Air temperature at 1.5 m | deg_C | tair | Temperature | K  | Same temperature variable, but units differ: CaSR uses Celsius, HRDPS uses Kelvin. |
| Forecast: Sea level pressure at surface | mb | atmpres | Pressure Reduced to MSL | Pa  | Same sea-level pressure concept, but units differ: CaSR uses millibars, HRDPS uses pascals. |
| Forecast: Surface incoming infrared flux at surface | W m**-2 | therm_rad | Downward Long-Wave Rad. Flux | W/m^2  | Incoming infrared flux corresponds to downward long-wave radiation. |
| Forecast: Quantity of precipitation model at surface | m | precip | Total Precipitation | kg/m^2  | Same precipitation-related variable, but units differ. For water, 1 kg/m² is approximately 1 mm water equivalent. |
| Forecast: Downward solar flux at surface | W m**-2 | solar | Downward Short-Wave Radiation Flux | W/m^2  | Downward solar flux corresponds to downward short-wave radiation. |
|Not Provided | | percentcloud | Total Cloud Cover | percent | This variable appears in HRDPS but was not included in the provided CaSR variable list.

```python
import xarray as xr
import numpy as np
import pandas as pd
import os

# 基础路径与配置
base_path = '/ocean/jqiu/Atmospheric_RDPS/2008_2024_Integrated/'
target_date = '2023-03-01' # 先截取 2023 年 3 月 1 日进行测试

# 目标变量名 : (原文件名, 原变量名)
var_mapping = {
    'solar':     ('Integrated_RDPS_P_FB_SFC_2008_2024.nc',  'CaSR_v3.2_P_FB_SFC'),
    'precip':    ('Integrated_RDPS_P_PR0_SFC_2008_2024.nc', 'CaSR_v3.2_P_PR0_SFC'),
    'therm_rad': ('Integrated_RDPS_P_FI_SFC_2008_2024.nc',  'CaSR_v3.2_P_FI_SFC'),
    'tair':      ('Integrated_RDPS_P_TT_1.5m_2008_2024.nc', 'CaSR_v3.2_P_TT_1.5m'),
    'qair':      ('Integrated_RDPS_P_HU_1.5m_2008_2024.nc', 'CaSR_v3.2_P_HU_1.5m'),
    'u_wind':    ('Integrated_RDPS_P_UUC_10m_2008_2024.nc', 'CaSR_v3.2_P_UUC_10m'),
    'atmpres':   ('Integrated_RDPS_P_PN_SFC_2008_2024.nc',  'CaSR_v3.2_P_PN_SFC'),
    'v_wind':    ('Integrated_RDPS_P_VVC_10m_2008_2024.nc', 'CaSR_v3.2_P_VVC_10m')
}

# 初始化目标 Dataset 与坐标
ds_out = xr.Dataset()

# 用 TT(温度) 文件作为参考来获取网格和时间信息
ref_file = os.path.join(base_path, 'Integrated_RDPS_P_TT_1.5m_2008_2024.nc')
ds_ref = xr.open_dataset(ref_file).sel(time=target_date)

y_size, x_size = ds_ref.rlat.size, ds_ref.rlon.size
time_size = ds_ref.time.size  # 获取实际的时间步数量，替代硬编码的 24

# 建立目标格式要求的坐标系
ds_out.coords['y'] = np.arange(y_size)
ds_out.coords['x'] = np.arange(x_size)
# 【修改点 1】直接提取原始数据的标准时间并赋给 time_counter
ds_out.coords['time_counter'] = ds_ref['time'].values 

# 重命名经纬度并赋值 (维度变为 y, x)
ds_out['nav_lat'] = (('y', 'x'), ds_ref['lat'].values)
ds_out['nav_lon'] = (('y', 'x'), ds_ref['lon'].values)
ds_out['nav_lat'].attrs['units'] = 'degrees_north'
ds_out['nav_lon'].attrs['units'] = 'degrees_east'

ds_ref.close()

# 循环读取并转换各变量
print(f"开始处理 {target_date} 的数据...")

for target_var, (fname, source_var) in var_mapping.items():
    fpath = os.path.join(base_path, fname)
    ds_in = xr.open_dataset(fpath).sel(time=target_date)
    
    # 提取数组
    data = ds_in[source_var].values
    
    # 执行单位转换
    if target_var == 'tair':
        data = data + 273.15        # deg_C -> K
        units = 'K'
    elif target_var in ['u_wind', 'v_wind']:
        data = data * 0.514444      # kts -> m/s
        units = 'm/s'
    elif target_var == 'atmpres':
        data = data * 100.0         # mb -> Pa
        units = 'Pa'
    elif target_var == 'precip':
        data = data * 1000.0        # m -> kg/m^2 (等效 1mm)
        units = 'kg/m^2'
    elif target_var in ['solar', 'therm_rad']:
        units = 'W/m^2'
    elif target_var == 'qair':
        units = 'kg/kg'
        
    # 存入新 Dataset，维度固定为 (time_counter, y, x)
    ds_out[target_var] = (('time_counter', 'y', 'x'), data)
    ds_out[target_var].attrs['units'] = units
    
    ds_in.close()

# 补充缺失的云覆盖率变量 (使用 time_size 动态适配时间步长度)
ds_out['percentcloud'] = (('time_counter', 'y', 'x'), np.zeros((time_size, y_size, x_size)))
ds_out['percentcloud'].attrs['units'] = 'percent'

# 强制规范时间变量的编码格式
# dtype 设为 float64 或 int64 防止秒数太大溢出（通常 float64 最稳妥）
out_encoding = {
    'time_counter': {
        'units': 'seconds since 1970-01-01 00:00:00',
        'calendar': 'standard',
        'dtype': 'float64'
    }
}

date_str = pd.to_datetime(target_date).strftime('%Ym%md%d')
output_filename = f'CaSR_y{date_str}.nc'

# 导出时传入 encoding 字典
ds_out.to_netcdf(
    output_filename, 
    unlimited_dims=('time_counter',),
    encoding=out_encoding
)

print(f"生成完毕: {output_filename}")
```

```python
import netCDF4 as nc
import numpy as np

# Define file paths
path_CaSR = '/home/jqiu/analysis-junqi/Analysis_Atmospheric_Forcing/Data_CaSR_Conversion/CaSR_y2008m01d01.nc'
path_HRDPS = '/results/forcing/atmospheric/GEM2.5/operational/ops_y2018m01d01.nc'

# Target variables for comparison
TARGET_VARS = [
    'time_counter', 'x', 'y', 
    'nav_lat', 'nav_lon', 
    'atmpres', 'percentcloud', 'precip', 
    'qair', 'solar', 'tair', 
    'therm_rad', 'u_wind', 'v_wind'
]

def get_var_info(dataset, var_name):
    """Extract attributes and sample data for a single variable, returning a dictionary."""
    if var_name not in dataset.variables:
        return {"shape": "Missing", "type": "Missing", "units": "Missing", "sample": "Missing"}
    
    var_obj = dataset.variables[var_name]
    info = {
        "shape": str(var_obj.shape),
        "type": str(var_obj.datatype),
        "units": getattr(var_obj, 'units', 'None')  # Display "None" if units attribute is missing
    }
    
    # Safely read a small amount of sample data
    if np.prod(var_obj.shape) > 0:
        try:
            if len(var_obj.shape) == 1:
                sample = var_obj[:3]
            elif len(var_obj.shape) == 2:
                sample = var_obj[0, :3]
            elif len(var_obj.shape) == 3:
                sample = var_obj[0, 0, :3]
            elif len(var_obj.shape) == 4:
                sample = var_obj[0, 0, 0, :3]
            else:
                sample = var_obj[...].flatten()[:3]
            
            # Format array: truncate floats to avoid long strings, take first 3 values
            if sample.dtype.kind == 'f':
                sample_str = "[" + ", ".join([f"{val:.6g}" for val in sample]) + ", ...]"
            else:
                sample_str = "[" + ", ".join([str(val) for val in sample]) + ", ...]"
                
            info["sample"] = sample_str
        except Exception:
            info["sample"] = "Read Error"
    else:
        info["sample"] = "Empty Array"
        
    return info

def compare_nc_files_as_table(file1, file2, vars_to_check):
    print("\nOpening files for side-by-side comparison...")
    try:
        ds1 = nc.Dataset(file1, 'r')
        ds2 = nc.Dataset(file2, 'r')
    except Exception as e:
        print(f"Failed to open files: {e}")
        return
    
    # Set table column widths
    col1_w = 14  # Variable Name
    col2_w = 10  # Attribute
    col3_w = 36  # File 1 Data
    col4_w = 36  # File 2 Data
    
    # Print table header
    sep_line = "-" * (col1_w + col2_w + col3_w + col4_w + 13)
    print(sep_line)
    print(f"| {'Variable':<{col1_w}} | {'Attribute':<{col2_w}} | {'CaSR (File 1)':<{col3_w}} | {'HRDPS (File 2)':<{col4_w}} |")
    print(sep_line)
    
    # Iterate through and compare each variable
    for var in vars_to_check:
        info1 = get_var_info(ds1, var)
        info2 = get_var_info(ds2, var)
        
        # First row: Variable Name and Shape
        print(f"| {var:<{col1_w}} | {'Shape':<{col2_w}} | {info1['shape']:<{col3_w}} | {info2['shape']:<{col4_w}} |")
        # Subsequent rows: Type, Units, Sample (Variable column left blank for clean look)
        print(f"| {'':<{col1_w}} | {'Type':<{col2_w}} | {info1['type']:<{col3_w}} | {info2['type']:<{col4_w}} |")
        print(f"| {'':<{col1_w}} | {'Units':<{col2_w}} | {info1['units']:<{col3_w}} | {info2['units']:<{col4_w}} |")
        print(f"| {'':<{col1_w}} | {'Sample':<{col2_w}} | {info1['sample']:<{col3_w}} | {info2['sample']:<{col4_w}} |")
        print(sep_line)

    ds1.close()
    ds2.close()

# Execute comparison
compare_nc_files_as_table(path_CaSR, path_HRDPS, TARGET_VARS)
```

```python
import xarray as xr
import numpy as np
import pandas as pd
import os
import time
from concurrent.futures import ProcessPoolExecutor, as_completed

def process_single_date(target_date, base_path, output_dir):
    """
    工作进程：处理单一日期的数据。
    (每个进程独立打开文件，避免多进程共享 NetCDF 句柄导致的崩溃)
    """
    date_str_out = target_date.strftime('%Ym%md%d')
    date_query = target_date.strftime('%Y-%m-%d')
    output_filename = os.path.join(output_dir, f'CaSR_y{date_str_out}.nc')

    # 断点续传：如果文件已存在，直接跳过
    if os.path.exists(output_filename):
        return f"跳过 (已存在): {date_query}"

    var_mapping = {
        'solar':     ('Integrated_RDPS_P_FB_SFC_2008_2024.nc',  'CaSR_v3.2_P_FB_SFC'),
        'precip':    ('Integrated_RDPS_P_PR0_SFC_2008_2024.nc', 'CaSR_v3.2_P_PR0_SFC'),
        'therm_rad': ('Integrated_RDPS_P_FI_SFC_2008_2024.nc',  'CaSR_v3.2_P_FI_SFC'),
        'tair':      ('Integrated_RDPS_P_TT_1.5m_2008_2024.nc', 'CaSR_v3.2_P_TT_1.5m'),
        'qair':      ('Integrated_RDPS_P_HU_1.5m_2008_2024.nc', 'CaSR_v3.2_P_HU_1.5m'),
        'u_wind':    ('Integrated_RDPS_P_UUC_10m_2008_2024.nc', 'CaSR_v3.2_P_UUC_10m'),
        'atmpres':   ('Integrated_RDPS_P_PN_SFC_2008_2024.nc',  'CaSR_v3.2_P_PN_SFC'),
        'v_wind':    ('Integrated_RDPS_P_VVC_10m_2008_2024.nc', 'CaSR_v3.2_P_VVC_10m')
    }

    try:
        ds_out = xr.Dataset()

        # 1. 提取网格基准和时间信息 (用当天的数据提取，耗时极短)
        ref_file = os.path.join(base_path, 'Integrated_RDPS_P_TT_1.5m_2008_2024.nc')
        with xr.open_dataset(ref_file) as ds_ref_full:
            ds_ref = ds_ref_full.sel(time=date_query)
            y_size, x_size = ds_ref.rlat.size, ds_ref.rlon.size
            lat_matrix = ds_ref['lat'].values
            lon_matrix = ds_ref['lon'].values
            time_values = ds_ref['time'].values  # 提取真实时间
            time_size = ds_ref.time.size         # 动态获取时间步数

        ds_out.coords['y'] = np.arange(y_size)
        ds_out.coords['x'] = np.arange(x_size)
        ds_out.coords['time_counter'] = time_values # 赋真实 datetime64 值
        
        ds_out['nav_lat'] = (('y', 'x'), lat_matrix)
        ds_out['nav_lon'] = (('y', 'x'), lon_matrix)
        ds_out['nav_lat'].attrs['units'] = 'degrees_north'
        ds_out['nav_lon'].attrs['units'] = 'degrees_east'

        # 2. 循环处理各变量
        for target_var, (fname, source_var) in var_mapping.items():
            fpath = os.path.join(base_path, fname)
            
            # 使用 with 语句确保读取后立刻释放文件锁
            with xr.open_dataset(fpath) as ds_in_full:
                day_slice = ds_in_full[source_var].sel(time=date_query).values

            # 物理单位转换
            if target_var == 'tair':
                data, units = day_slice + 273.15, 'K'
            elif target_var in ['u_wind', 'v_wind']:
                data, units = day_slice * 0.514444, 'm/s'
            elif target_var == 'atmpres':
                data, units = day_slice * 100.0, 'Pa'
            elif target_var == 'precip':
                data, units = day_slice * 1000.0, 'kg/m^2'
            elif target_var in ['solar', 'therm_rad']:
                data, units = day_slice, 'W/m^2'
            elif target_var == 'qair':
                data, units = day_slice, 'kg/kg'
                
            ds_out[target_var] = (('time_counter', 'y', 'x'), data)
            ds_out[target_var].attrs['units'] = units

        # 3. 添加云覆盖 (动态适配时间维大小)
        ds_out['percentcloud'] = (('time_counter', 'y', 'x'), np.zeros((time_size, y_size, x_size), dtype=np.float32))
        ds_out['percentcloud'].attrs['units'] = 'percent'

        # 4. 导出 (添加 unlimited_dims 和 encoding)
        out_encoding = {
            'time_counter': {
                'units': 'seconds since 1970-01-01 00:00:00',
                'calendar': 'standard',
                'dtype': 'float64'
            }
        }
        
        ds_out.to_netcdf(
            output_filename, 
            format='NETCDF4', 
            unlimited_dims=('time_counter',), 
            encoding=out_encoding
        )
        return f"成功生成: {date_query}"

    except Exception as e:
        return f"失败 {date_query} 报错: {str(e)}"

def run_parallel_conversion(base_path, output_dir, start_date='2008-01-01', end_date='2024-12-31', max_workers=8):
    os.makedirs(output_dir, exist_ok=True)
    date_range = pd.date_range(start=start_date, end=end_date, freq='D')
    total_days = len(date_range)
    
    print(f"开始并行处理 {total_days} 天的数据，使用 {max_workers} 个进程...")
    start_time = time.time()

    success_count = 0
    # 使用进程池
    with ProcessPoolExecutor(max_workers=max_workers) as executor:
        # 提交所有任务到进程池
        futures = {
            executor.submit(process_single_date, date, base_path, output_dir): date 
            for date in date_range
        }
        
        # 实时获取完成状态
        for i, future in enumerate(as_completed(futures), 1):
            result = future.result()
            
            if "成功" in result or "跳过" in result:
                success_count += 1
            
            # 每完成 50 个打印一次进度，防止日志刷屏
            if i % 50 == 0 or i == total_days:
                elapsed = time.time() - start_time
                speed = i / elapsed
                print(f"[{i}/{total_days}] {result} (平均速度: {speed:.2f} 天/秒)")

    total_time = time.time() - start_time
    print(f"\n全部完成！耗时: {total_time/60:.2f} 分钟。成功处理: {success_count}/{total_days} 天。")

if __name__ == "__main__":
    SOURCE_PATH = '/ocean/jqiu/Atmospheric_RDPS/2008_2024_Integrated/'
    OUTPUT_PATH = '/ocean/jqiu/CaSR/'
    
    # max_workers 建议设置为 CPU 核心数的 50%~80%
    run_parallel_conversion(
        base_path=SOURCE_PATH, 
        output_dir=OUTPUT_PATH,
        start_date='2008-01-01', 
        end_date='2024-12-31',
        max_workers=6  # <--- 这里可以根据服务器性能调整
    )
```
